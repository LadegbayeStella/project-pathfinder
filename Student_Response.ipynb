## Students Response Analysis

The dataset for the analysis is from the students survey sent out for the Mentorship Platform research. The goal of the analysis is to gather insights from the data and build visualisations to commuinicate the insights. The insights will aid in tailoring the platform to suit users need.  

```python
#import the needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the needed libraries and load in the dataset
import pandas as pd

students = pd.read_csv('Responses - Students.csv')

#exlore the dataset
students.head()
```

```python
#drop the redundant columns 
students.drop(columns = ['Timestamp','What Country do you reside in?', 'Have you ever faced challenges in finding a mentor who aligns with your career interests?', 'What challenges have you faced while getting a mentor?',
                         '\nDo you have access to technology and the internet for educational purposes?', 'How important is it for you to have access to teachers and counselors through an educational networking platform?', 'How likely are you to participate in webinars or training sessions related to your career interests?',
                         'How do you think a platform like the one described could improve your career development? List all applicable ideas.'], inplace = True)
```

```python
#change column names to make it more readable
students.columns = ['Age', 'Gender', 'State', 'Future Career', 'Mentorship importance',
                    'Career Assistance', 'Assistance Frequency', 'Mentorship Value', 'Current mentors',
                    'Mentorship Frequency', 'Likely Platform Usage', 'Platform features']
students.head()
```

```python
students.info()
```

```python
students.describe()
```

```python
#replace the inconsistent entries in the mentorship frequency column
students['Mentorship Frequency'].replace('None','Nil',inplace = True)
students['Mentorship Frequency'].replace('Nil ','Nil',inplace = True)
```

```python
students.head()
```

```python
#plot  a bar chart for all categorical columns to visualise the distribution
categorical_columns1 = ['Age', 'Gender', 'Career Assistance']
categorical_columns2 = ['Assistance Frequency', 'Mentorship Value',
                      'Current mentors'] 
categorical_columns3 = ['Mentorship Frequency', 'Likely Platform Usage']

for col in categorical_columns1:
    plt.figure(figsize = (8,5))
    sns.countplot(data = students, x= col)
    plt.title(f'Count of Students by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#plot  a bar chart for all categorical columns to visualise the distribution
for col in categorical_columns2:
    plt.figure(figsize = (10,8))
    sns.countplot(data = students, x= col)
    plt.title(f'Count of Students by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#plot  a bar chart for all categorical columns to visualise the distribution
for col in categorical_columns3:
    plt.figure(figsize = (10,8))
    sns.countplot(data = students, x= col)
    plt.title(f'Count of Students by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

### Observations

Based on the visualisations, these are the following observations;
- Our survey has more of students wth age range 12-18 years with more females(24) than males(13). 
- A good number of students (27) have received prior assistance in relation to their career aspirations.
